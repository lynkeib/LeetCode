```
public class Solution {
    public bool AreSentencesSimilarTwo(string[] words1, string[] words2, IList<IList<string>> pairs) {
        
        if(words1.Length != words2.Length){
            return false;
        }
        
        var map = new Dictionary<string, int>();
        var count = 0;
        
        foreach(var w in words1){
            if(!map.ContainsKey(w)){
                map[w] = count;
                count++;
            }
        }
        foreach(var w in words2){
            if(!map.ContainsKey(w)){
                map[w] = count;
                count++;
            }
        }
        foreach(var l in pairs){
            var w1 = l[0];
            var w2 = l[1];
            if(!map.ContainsKey(w1)){
                map[w1] = count;
                count++;
            }
            if(!map.ContainsKey(w2)){
                map[w2] = count;
                count++;
            }
        }
        
        var n = count;
        
        var uf = new int[n];
        for(int i = 0; i < n; i++) uf[i] = i;
        
        foreach(var l in pairs){
            var w1 = l[0];
            var w2 = l[1];
            
            var n1 = map[w1];
            var n2 = map[w2];
            
            var p_n1 = find(uf, n1);
            var p_n2 = find(uf, n2);
            
            if(p_n1 != p_n2){
                if(p_n2 < p_n1){
                    var temp = p_n2;
                    p_n2 = p_n1;
                    p_n1 = temp;
                }
                uf[p_n2] = p_n1;
            }
        }
        
        for(int i = 0; i < words1.Length; i++){
            var w1 = words1[i];
            var w2 = words2[i];
            
            var n1 = map[w1];
            var n2 = map[w2];
            
            var p_n1 = find(uf, n1);
            var p_n2 = find(uf, n2);
            
            if(p_n1 != p_n2){
                return false;
            }
        }
        
        return true;
        
    }
    
    public int find(int[] uf, int node){
        if(node != uf[node]){
            node = find(uf, uf[node]);
        }
        return node;
    }
}
```
