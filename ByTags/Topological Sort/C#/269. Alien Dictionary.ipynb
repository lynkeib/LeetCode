```
public class Solution {

    bool error = false;

    public string AlienOrder (string[] words) {
        var map = new Dictionary<char, HashSet<char>> ();
        var indegree = new int[26];
        buildGraph (map, words, indegree);
        if(error){
            return "";
        }
        return bfs (map, indegree);
    }

    public void buildGraph (Dictionary<char, HashSet<char>> map, string[] words, int[] indegree) {
        foreach (var word in words) {
            foreach (var c in word) {
                if (!map.ContainsKey (c)) {
                    map[c] = new HashSet<char> ();
                }
            }
        }
        for (int i = 1; i < words.Length; i++) {
            var first = words[i - 1];
            var second = words[i];
            var len = Math.Min (first.Length, second.Length);
            var counter = 0;
            for (int j = 0; j < len; j++) {
                if (first[j] != second[j]) {
                    var OUT = first[j];
                    var IN = second[j];
                    if (!map[OUT].Contains (IN)) {
                        map[OUT].Add (IN);
                        indegree[IN - 'a']++;
                    }
                    break;
                }
                counter++;
            }
            if(len == second.Length && counter == len && counter != first.Length){
                error = true;
            }
            if(error){
                return;
            }
        }
        
    }

    public string bfs (Dictionary<char, HashSet<char>> map, int[] indegree) {
        var sb = new StringBuilder ();
        var queue = new Queue<char> ();
        var len = map.Keys.Count;
        foreach (var key in map.Keys) {
            if (indegree[key - 'a'] == 0) {
                sb.Append (key);
                queue.Enqueue (key);
            }
        }

        while (queue.Count != 0) {
            var OUT = queue.Dequeue ();
            if(map[OUT].Count == 0) continue;
            foreach (var IN in map[OUT]) {
                indegree[IN - 'a']--;
                if (indegree[IN - 'a'] == 0) {
                    sb.Append (IN);
                    queue.Enqueue (IN);
                }
            }
        }
        return sb.Length == len ? sb.ToString () : "";
    }
}
```
