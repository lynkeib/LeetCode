```
public class Solution {
    
    public int[] FindOrder(int numCourses, int[][] prerequisites) {
        
        Dictionary<int, HashSet<int>> graph = new Dictionary<int, HashSet<int>>();
        int[] indegree = new int[numCourses];
        
        List<int> res = new List<int>();
        
        // init graph and indegree
        foreach(int[] row in prerequisites){
            int c = row[0], pre = row[1];
            if(!graph.ContainsKey(pre)){
                graph[pre] = new HashSet<int>();
            }
            graph[pre].Add(c);
            indegree[c] += 1;
        }
        Queue<int> queue = new Queue<int>();
        for(int index=0;index<numCourses;index++){
            if(indegree[index]==0){
                queue.Enqueue(index);
            }
        }
        while(queue.Count != 0){
            int n = queue.Dequeue();
            dfs(n, ref graph, ref queue, ref indegree, ref res);
        }
        foreach(int d in indegree){
            if(d != 0){
                return new int[0];
            }
        }
        return res.ToArray();
    }
    
    public void dfs(int node, ref Dictionary<int, HashSet<int>> graph, ref Queue<int> queue, ref int[] indegree, ref List<int> res){
        res.Add(node);
        if(!graph.ContainsKey(node)){
            return;
        }
        Queue<int> next_queue = new Queue<int>();
        foreach(int n in graph[node]){
            indegree[n] -= 1;
            if(indegree[n] == 0){
                next_queue.Enqueue(n);
            }
        }
        graph.Remove(node);
        foreach(int n in next_queue){
            dfs(n, ref graph, ref next_queue, ref indegree, ref res);
        }
        return;
    }
}
```
