```python
class Solution(object):
    def findMedianSortedArrays(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: float
        """
        # say we have y numbers in nums1 and x numbers in nums2
        # 
        
        length1 = len(nums1)
        length2 = len(nums2)
        if length1 > length2:
            nums1, nums2 = nums2, nums1
            length1, length2 = length2, length1
            # from C0 to Ck-1, there are k elements
            # in which contains m1 elements from nums1 and m2 elements from nums2
            # here we need to consider how to use m1
            # say m1 == 0, then we should not use any elements in nums1
            # if m1 == 0, then m2 = k - m1 (say k = 6), then m2 = 6
            # what does the m2 mean here?
            # m2 means we should use 6 elements in nums2 to construct k
            # However, is it ture that the last element of nums2 (with index of 5) is the middle one?
            # so here we nned to chech whether the left is right here
            # how to check?
            # since we didn't select anything in nums1
            # so the first element in nums1 should be greater than the the 6th element in nums2
            # because if nums1 is smaller than or equal to nums2, then we should include that element in the k elements
            # by knowing this, we can then begin
            
        k = (length1 + length2 + 1) // 2
        left, right = 0, length1
        while left < right:
            mid1 = left + (right - left) // 2
            mid2 = k - mid1
            if nums1[mid1] < nums2[mid2 - 1]:
                left = mid1 + 1
            else:
                right = mid1
                
                
        mid1 = max(nums1[left - 1] if left > 0 else -float('inf'),
                   nums2[(k - left) - 1] if (k - left) > 0 else -float('inf'))
        
        if (length1 + length2) % 2 != 0:
            return mid1
        
        mid2 = min(nums1[left] if left < length1 else float('inf'),
                   nums2[(k - left)] if (k - left) < length2 else float('inf'))
        
        return (mid1 + mid2) * 1.0 / 2
        # print(k)
        # left here means in nums1, we should select 3 elements to construct k
        # so by having say 8 elements ot select, we should select 5 elements in nums2
        # then which one is the last one?
        # if (length1 + length2) % 2 != 0:
        #     # if we did not select any element in nums1
        #     # meaning all the elemtns are come from nums2
        #     if left == 0:
        #         return nums2[k - 1]
        #     elif left == k:
        #         return nums1[left - 1]
        #     else:
        #         return max(nums1[left - 1], nums2[(k - left) - 1])
        # else:
        #     # same, if we did not select any element in nums1
        #     # print('left', left)
        #     # print(length1 + length2)
        #     # print('k', k)
        #     if left == 0:
        #         mid1 = nums2[k - 1]
        #         print(mid1)
        #         if k == length2:
        #             mid2 = nums1[0]
        #         else:
        #             mid2 = min(nums1[0], nums2[k])
        #         return (mid1 + mid2) * 1.0 / 2
        #     elif left == length1:
        #         mid1 = nums1[left - 1]
        #         mid2 = nums2[0]
        #         return (mid1 + mid2) * 1.0 / 2
        #     else:
        #         # print((k - left) - 1)
        #         mid1 = max(nums1[left - 1], nums2[(k - left) - 1])
        #         print(mid1)
        #         # if left is the last element in nums1
        #         mid2 = min(nums1[left], nums2[k - left])
        #         return (mid1 + mid2) * 1.0 / 2
```
