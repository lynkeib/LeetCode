```
public class Solution {
    public int Divide (int dividend, int divisor) {
        if (divisor == 0) {
            return int.MaxValue;
        }
        bool is_positive = true;
        if (dividend > 0) {
            is_positive = !is_positive;
            dividend = -dividend;
        }
        if (divisor > 0) {
            is_positive = !is_positive;
            divisor = -divisor;
        }

        int counter = 1;
        int half = int.MinValue / 2;
        var map = new List<int> ();
        var poweroftwo = new List<int> ();
        while (dividend <= divisor) {
            map.Add (divisor);
            poweroftwo.Add (counter);
            if (divisor < half) {
                break;
            }
            divisor += divisor;
            counter += counter;
        }

        int res = 0;
        for (int i = map.Count - 1; i >= 0; i--) {
            if (map[i] >= dividend) {
                res -= poweroftwo[i];
                dividend -= map[i];
            }
        }
        if (is_positive && res == int.MinValue) {
            return int.MaxValue;
        }
        return is_positive? - res : res;
    }
}
```

# Bit Manipulation

```
public class Solution {
    public int Divide (int dividend, int divisor) {
        if (divisor == 0) {
            return int.MaxValue;
        }
        bool is_positive = true;
        if (dividend > 0) {
            is_positive = !is_positive;
            dividend = -dividend;
        }
        if (divisor > 0) {
            is_positive = !is_positive;
            divisor = -divisor;
        }
        
        int original_divisor = divisor;
        int counter = 1;
        int half = int.MinValue / 2;
        while (dividend <= divisor + divisor) {
            if (divisor < half) {
                break;
            }
            divisor <<= 1;
            counter <<= 1;
        }
        int res = 0;
        while (dividend <= original_divisor) {
            if(dividend <= divisor){
                res -= counter;
                dividend -= divisor;
            }
            counter >>= 1;
            divisor >>= 1;
        }
        if (is_positive && res == int.MinValue) {
            return int.MaxValue;
        }
        return is_positive? - res : res;
    }
}
```
