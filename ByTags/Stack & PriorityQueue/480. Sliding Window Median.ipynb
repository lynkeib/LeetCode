```python
class Median():
    def __init__(self, nums, k):
        self.maxHeap, self.minHeap = [], []
        self.invalidIndex = set([])
        self.maxInvalid, self.minInvalid = 0, 0
        self.maxIndex, self.minIndex = set([]), set([])
        self.k = k
        self.nums = nums
    
    def getMedian(self):
        validMax = len(self.maxHeap) - self.maxInvalid
        validMin = len(self.minHeap) - self.minInvalid
        if (validMax + validMin) % 2 != 0:
            return -self.maxHeap[0][0]
        else:
            return (-self.maxHeap[0][0] + self.minHeap[0][0]) / 2.0
    
    def add(self, index):
        invalidIdx = index - self.k
        value = self.nums[index]
        
        self.invalidIndex.add(invalidIdx)
        if invalidIdx in self.maxIndex:
            self.maxInvalid += 1
        elif invalidIdx in self.minIndex:
            self.minInvalid += 1

        if len(self.maxHeap) - self.maxInvalid == 0 or -self.maxHeap[0][0] >= value:
            heapq.heappush(self.maxHeap, (-value, index))
            self.maxIndex.add(index)
        else:
            heapq.heappush(self.minHeap, (value, index))
            self.minIndex.add(index)
        
            
        while (len(self.maxHeap) - self.maxInvalid > len(self.minHeap) - self.minInvalid + 1) or (len(self.maxHeap) - self.maxInvalid < len(self.minHeap) - self.minInvalid) or (len(self.maxHeap) != 0 and self.maxHeap[0][1] in self.invalidIndex) or (len(self.minHeap) != 0 and self.minHeap[0][1] in self.invalidIndex) :
            
            while len(self.maxHeap) != 0 and self.maxHeap[0][1] in self.invalidIndex:
                value, index = heapq.heappop(self.maxHeap)
                self.maxIndex.remove(index)
                self.invalidIndex.remove(index)
                self.maxInvalid -= 1
            while len(self.minHeap) != 0 and self.minHeap[0][1] in self.invalidIndex:
                value, index = heapq.heappop(self.minHeap)
                self.minIndex.remove(index)
                self.invalidIndex.remove(index)
                self.minInvalid -= 1
            if len(self.maxHeap) - self.maxInvalid > len(self.minHeap) - self.minInvalid + 1:
                value, index = heapq.heappop(self.maxHeap)
                heapq.heappush(self.minHeap, (-value, index))
                self.maxIndex.remove(index)
                self.minIndex.add(index)
            if len(self.maxHeap) - self.maxInvalid < len(self.minHeap) - self.minInvalid:
                value, index = heapq.heappop(self.minHeap)
                heapq.heappush(self.maxHeap, (-value, index))
                self.minIndex.remove(index)
                self.maxIndex.add(index)

class Solution(object):
    def medianSlidingWindow(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: List[float]
        """
        median = Median(nums, k)
        for i in range(k - 1):
            median.add(i)
        res = []
        for i in range(k - 1, len(nums)):
            median.add(i)
            # print(median.maxHeap,  median.minHeap)
            # print(median.maxInvalid,median.minInvalid)
            # print(median.maxIndex, median.minIndex)
            # print(i, median.invalidIndex)
            # print('-----------')
            res.append(median.getMedian())
        return res
        
```
