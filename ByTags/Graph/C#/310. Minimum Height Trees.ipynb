# First Try

```
public class Solution {
    public IList<int> FindMinHeightTrees (int n, int[][] edges) {
        IList<int> res = new List<int> ();
        if(edges.Count() == 0){
            for(int i=0;i<n;i++) res.Add(i);
            return res;
        }

        Dictionary<int, HashSet<int>> graph = new Dictionary<int, HashSet<int>> ();
        Dictionary<int, Dictionary<int, int>> height = new Dictionary<int, Dictionary<int, int>> ();

        foreach (int[] pair in edges) {
            
            int n1 = pair[0], n2 = pair[1];
            if (!graph.ContainsKey (n1)) graph[n1] = new HashSet<int> ();
            if (!graph.ContainsKey (n2)) graph[n2] = new HashSet<int> ();
            graph[n1].Add (n2);
            graph[n2].Add (n1);
        }
        foreach (int node in graph.Keys) {
            HashSet<int> visited = new HashSet<int> ();
            calheight (-1, node, ref graph, ref height, ref visited);
        }
        int min_height = int.MaxValue;
        foreach (int node in height[-1].Keys) {
            min_height = Math.Min (min_height, height[-1][node]);
        }
        foreach (int node in height[-1].Keys) {
            if (height[-1][node] == min_height) {
                res.Add (node);
            }
        }
        return res;
    }

    public int calheight (
        int parent,
        int node,
        ref Dictionary<int, HashSet<int>> graph,
        ref Dictionary<int, Dictionary<int, int>> height,
        ref HashSet<int> visited) {
        if (height.ContainsKey (parent)) {
            if (height[parent].ContainsKey (node)) {
                return height[parent][node];
            }
        }
        visited.Add (node);
        if (!height.ContainsKey (parent)) height[parent] = new Dictionary<int, int> ();

        int max_height = 0;
        foreach (int n in graph[node]) {
            if (!visited.Contains (n)) {
                max_height = Math.Max (max_height, calheight (node, n, ref graph, ref height, ref visited));
            }
        }
        height[parent][node] = max_height + 1;
        return height[parent][node];
    }
}
```

# Faster

```
public class Solution {
    public IList<int> FindMinHeightTrees(int n, int[][] edges) {
        List<int> res = new List<int>();
        if(n==1){
            res.Add(0);
            return res;
        }
        Dictionary<int, HashSet<int>> graph = new Dictionary<int, HashSet<int>>();
        foreach(int[] pair in edges){
            int n1 = pair[0], n2 = pair[1];
            if(!graph.ContainsKey(n1)) graph[n1] = new HashSet<int>();
            if(!graph.ContainsKey(n2)) graph[n2] = new HashSet<int>();
            graph[n1].Add(n2);
            graph[n2].Add(n1);
        }
        HashSet<int> leaves = new HashSet<int>();
        foreach(int node in graph.Keys) if(graph[node].Count == 1) leaves.Add(node);
 
        while(n>2){
            HashSet<int> next_leaves = new HashSet<int>();
            foreach(int node in leaves){
                foreach(int nei in graph[node]){
                    graph[nei].Remove(node);
                    if(graph[nei].Count == 1) next_leaves.Add(nei);
                }
                n -= 1;
            }
            leaves = next_leaves;
        }

        return leaves.ToList();
    }
}
```
