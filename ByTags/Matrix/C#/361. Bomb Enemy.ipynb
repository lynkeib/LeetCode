```
public class Solution {

    int[] rows;
    int[] cols;

    public int MaxKilledEnemies (char[][] grid) {
        int nrows = grid.Length;
        if (nrows == 0) {
            return 0;
        }
        int ncols = grid[0].Length;

        rows = new int[nrows];
        cols = new int[ncols];

        for (int row = 0; row < nrows; row++) {
            for (int col = 0; col < ncols; col++) {
                if (grid[row][col] == 'W') {
                    rows[row] = -1;
                    cols[col] = -1;
                } else if (grid[row][col] == 'E') {
                    if (rows[row] != -1) {
                        rows[row]++;
                    }
                    if (cols[col] != -1) {
                        cols[col]++;
                    }
                }
            }
        }
        var res = 0;
        for (int row = 0; row < nrows; row++) {
            for (int col = 0; col < ncols; col++) {
                if (grid[row][col] == '0') {
                    var total = 0;
                    if (rows[row] != -1) {
                        total += rows[row];
                    } else {
                        // left
                        var curr_col = col;
                        curr_col--;
                        while (curr_col >= 0) {
                            if (grid[row][curr_col] == 'E') {
                                total++;
                            } else if (grid[row][curr_col] == 'W') {
                                break;
                            }
                            curr_col--;
                        }
                        // right
                        curr_col = col;
                        curr_col++;
                        while (curr_col < ncols) {
                            if (grid[row][curr_col] == 'E') {
                                total++;
                            } else if (grid[row][curr_col] == 'W') {
                                break;
                            }
                            curr_col++;
                        }
                    }

                    if (cols[col] != -1) {
                        total += cols[col];
                    } else {
                        // up
                        var curr_row = row;
                        curr_row--;
                        while (curr_row >= 0) {
                            if (grid[curr_row][col] == 'E') {
                                total++;
                            } else if (grid[curr_row][col] == 'W') {
                                break;
                            }
                            curr_row--;
                        }
                        // right
                        curr_row = row;
                        curr_row++;
                        while (curr_row < nrows) {
                            if (grid[curr_row][col] == 'E') {
                                total++;
                            } else if (grid[curr_row][col] == 'W') {
                                break;
                            }
                            curr_row++;
                        }
                    }

                    res = Math.Max(res, total);
                }

            }
        }
        return res;
    }
}
```
