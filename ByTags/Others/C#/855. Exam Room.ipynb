```
public class ExamRoom {

    int[] seats;
    int n;
    int count = 0;
    
    public ExamRoom(int N) {
        seats = new int[N + 2];
        seats[0] = 1;
        seats[N + 1] = 1;
        n = N;
    }
    
    public int Seat() {
        count++;
        // first add
        if(count == 0){
            seats[1] = 1;
            return 0;
        }
        
        int max_dist = -1;
        int res = -1;
        
        int curr = 0;
        while(curr < seats.Length){
            int prev = curr;
            curr++;
            while(curr < seats.Length && seats[curr] != 1){
                curr++;
            }
            int curr_dist = curr - prev;
            // adjacent
            if(curr_dist == 1){
                continue;
            }
            // start to last
            if(curr_dist == n){
                if(prev == 0){
                    res = 0;
                }else{
                    res = n - 1;
                }
                break;
            }
            // if seats 0 is empty
            if(prev == 0 && curr_dist - 1 > max_dist){
                max_dist = curr_dist - 1;
                res = 0;
            }
            // if seats last is empty
            else if(curr == n + 1 && curr_dist - 1 > max_dist){
                max_dist = curr_dist - 1;
                res = n - 1;
            }
            // normal cases
            else {
                int index = prev + (curr - prev) / 2;
                curr_dist = index - prev;
                if(curr_dist > max_dist){
                    max_dist = curr_dist;
                    res = index - 1;
                }
            }
        }
        seats[res + 1] = 1;
        return res;
    }
    
    public void Leave(int p) {
        seats[p + 1] = 0;
        count--;
        return;
    }
}

/**
 * Your ExamRoom object will be instantiated and called as such:
 * ExamRoom obj = new ExamRoom(N);
 * int param_1 = obj.Seat();
 * obj.Leave(p);
 */
```
