```
public class ExamRoom {

    int[] seats;
    int n;
    int count = 0;
    
    public ExamRoom(int N) {
        seats = new int[N + 2];
        seats[0] = 1;
        seats[N + 1] = 1;
        n = N;
    }
    
    public int Seat() {
        count++;
        // first add
        if(count == 0){
            seats[1] = 1;
            return 0;
        }
        
        int max_dist = -1;
        int res = -1;
        
        int curr = 0;
        while(curr < seats.Length){
            int prev = curr;
            curr++;
            while(curr < seats.Length && seats[curr] != 1){
                curr++;
            }
            int curr_dist = curr - prev;
            // adjacent
            if(curr_dist == 1){
                continue;
            }
            // start to last
            if(curr_dist == n){
                if(prev == 0){
                    res = 0;
                }else{
                    res = n - 1;
                }
                break;
            }
            // if seats 0 is empty
            if(prev == 0 && curr_dist - 1 > max_dist){
                max_dist = curr_dist - 1;
                res = 0;
            }
            // if seats last is empty
            else if(curr == n + 1 && curr_dist - 1 > max_dist){
                max_dist = curr_dist - 1;
                res = n - 1;
            }
            // normal cases
            else {
                int index = prev + (curr - prev) / 2;
                curr_dist = index - prev;
                if(curr_dist > max_dist){
                    max_dist = curr_dist;
                    res = index - 1;
                }
            }
        }
        seats[res + 1] = 1;
        return res;
    }
    
    public void Leave(int p) {
        seats[p + 1] = 0;
        count--;
        return;
    }
}

/**
 * Your ExamRoom object will be instantiated and called as such:
 * ExamRoom obj = new ExamRoom(N);
 * int param_1 = obj.Seat();
 * obj.Leave(p);
 */
```

```
public class ExamRoom {

    List<int> seats = new List<int>();
    int n;
    
    public ExamRoom(int N) {
        n = N;
    }
    
    public int Seat() {
        if(seats.Count == 0){
            seats.Add(0);
            return 0;
        }
        // start and end
        int max_dist = Math.Max(seats[0], n - 1 - seats[seats.Count - 1]);
        // middle
        for(int i = 0; i < seats.Count - 1; i++){
            int curr_dist = (seats[i + 1] - seats[i]) / 2;
            max_dist = Math.Max(max_dist, curr_dist);
        }
        // if start is the longest
        if(seats[0] == max_dist){
            seats.Insert(0, 0);
            return 0;
        }
        // find this dist
        for(int i = 0; i < seats.Count - 1; i++){
            int curr_dist = (seats[i + 1] - seats[i]) / 2;
            if(curr_dist == max_dist){
                int index = (seats[i + 1] + seats[i]) / 2;
                seats.Insert(i + 1, index);
                return index;
            }
        }
        // then must be the last one
        seats.Insert(seats.Count, n-1);
        return n - 1;
    }
    
    public void Leave(int p) {
        for(int i = 0; i < seats.Count; i++){
            if(seats[i] == p){
                seats.RemoveAt(i);
            }
        }
        return ;
    }
}

/**
 * Your ExamRoom object will be instantiated and called as such:
 * ExamRoom obj = new ExamRoom(N);
 * int param_1 = obj.Seat();
 * obj.Leave(p);
 */
```
