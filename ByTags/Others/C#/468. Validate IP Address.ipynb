```
public class Solution {
    public string ValidIPAddress(string IP) {
        if(IP.Contains('.')){
            if(IPv4(IP)){
                return "IPv4";
            }
        }else if(IP.Contains(':')){
            if(IPv6(IP)){
                return "IPv6";
            }
        }
        return "Neither";
    }
    
    public bool IPv4(string IP){
        if(IP.Contains(':')){
            return false;
        }
        string[] ips = IP.Split('.');
        if(ips.Length != 4){
            return false;
        }
        foreach(var ip in ips){
            if(!checkIPv4(ip)){
                return false;
            }
        }
        return true;
    }
    
    public bool checkIPv4(string ip){
        // empty
        if(ip == ""){
            return false;
        }
        // too long
        if(ip.Length > 3){
            return false;
        }
        // leading 0
        if(ip.Length != 1 && ip[0] == '0'){
            return false;
        }
        // has letters?
        foreach(char c in ip){
            if(c < '0' || c > '9'){
                return false;
            }
        }
        int ipNum = Int32.Parse(ip);
        return ipNum >=0 && ipNum <= 255;
    }
    
    public bool IPv6(string IP){
        if(IP.Contains('.')){
            return false;
        }
        string[] ips = IP.Split(':');
        if(ips.Length != 8){
            return false;
        }
        foreach(var ip in ips){
            // Console.WriteLine(ip);
            if(!checkIPv6(ip)){
                return false;
            }
        }
        return true;
    }
    
    public bool checkIPv6(string ip){
        // empty
        if(ip == ""){
            return false;
        }
        // extra leading zeros 
        if(ip.Length > 4){
            return false;
        }
        // else
        foreach(var c in ip){
            if(c >= '0' && c <= '9'){
                continue;
            }else{
                var C = Char.ToUpper(c);
                if(C < 'A' || C > 'F'){
                    return false;
                }
            }
        }
        return true;
    }
}
```
