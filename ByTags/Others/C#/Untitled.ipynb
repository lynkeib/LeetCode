```
public class Solution {
    public IList<IList<int>> PacificAtlantic (int[][] matrix) {
        var res = new List<IList<int>> ();
        if(matrix.Length == 0 || matrix[0].Length == 0){
            return res;
        }
        int rows = matrix.Length, cols = matrix[0].Length;
        var pacific = new bool[rows, cols];
        var atlantic = new bool[rows, cols];
        for (int row = 0; row < rows; row++) {
            dfs (matrix, pacific, row, 0, int.MinValue);
            dfs (matrix, atlantic, row, cols - 1, int.MinValue);
        }
        for (int col = 0; col < cols; col++) {
            dfs (matrix, pacific, 0, col, int.MinValue);
            dfs (matrix, atlantic, rows - 1, col, int.MinValue);
        }
        
        for (int row = 0; row < rows; row++) {
            for (int col = 0; col < cols; col++) {
                if(pacific[row, col] && atlantic[row, col]){
                    res.Add(new List<int>{row, col});
                }
            }
        }
        return res;
    }

    public void dfs (int[][] matrix, bool[, ] map, int row, int col, int height) {
        int rows = matrix.Length, cols = matrix[0].Length;
        if (row < 0 || row >= rows || col < 0 || col >= cols || matrix[row][col] < height || map[row, col]) {
            return;
        }
        map[row, col] = true;

        dfs (matrix, map, row + 1, col, matrix[row][col]);
        dfs (matrix, map, row - 1, col, matrix[row][col]);
        dfs (matrix, map, row, col + 1, matrix[row][col]);
        dfs (matrix, map, row, col - 1, matrix[row][col]);
        return;
    }
}
```

```
public class Solution {
    public IList<IList<int>> PacificAtlantic (int[][] matrix) {
        var dp = new Dictionary<int, Dictionary<int, bool[]>> ();
        var res = new List<IList<int>> ();
        var visited = new HashSet<string>();
        var finished =  new HashSet<string>();
        for (int row = 0; row < matrix.Length; row++) {
            for (int col = 0; col < matrix[0].Length; col++) {
                var c = dfs (matrix, row, col, ref dp, ref visited, ref finished);
                if (c[0] && c[1]) {
                    res.Add (new List<int> { row, col });
                    // Console.WriteLine(res);
                }
            }
        }
        return res;
    }

    public bool[] dfs (int[][] matrix, int row, int col, ref Dictionary<int, Dictionary<int, bool[]>> dp, ref HashSet<string> visited, ref HashSet<string> finished) {
        var key = $"{row}{col}";
        if (dp.ContainsKey (row) && dp[row].ContainsKey (col) && finished.Contains(key)) {
            return dp[row][col];
        }

        

        var res = new bool[2];
        // top or left
        if (row == 0 || col == 0) {
            res[0] = true;
        }
        // bottom or right
        if (row == matrix.Length - 1 || col == matrix[0].Length - 1) {
            res[1] = true;
        }
        if(res[0] && res[1]){
            dp[row][col] = res;
            finished.Add(key);
            return res;
        }
        // if has upper and upper is equal-height or lower
        if (row != 0 && matrix[row][col] >= matrix[row - 1][col] && !visited.Contains($"{row - 1}{col}")) {
            visited.Add(key);
            var upper = dfs (matrix, row - 1, col, ref dp, ref visited, ref finished);
            visited.Remove(key);
            res[0] = res[0] || upper[0];
            res[1] = res[1] || upper[1];
            if (res[0] && res[1]) {
                dp[row][col] = res;
                finished.Add(key);
                return res;
            }
        }
        // if has lower and lower is equal-height or lower
        if (row != matrix.Length - 1 && matrix[row][col] >= matrix[row + 1][col] && !visited.Contains($"{row + 1}{col}")) {
            visited.Add(key);
            var lower = dfs (matrix, row + 1, col, ref dp, ref visited, ref finished);
            visited.Remove(key);
            res[0] = res[0] || lower[0];
            res[1] = res[1] || lower[1];
            if (res[0] && res[1]) {
                dp[row][col] = res;
                finished.Add(key);
                return res;
            }
        }
        // if has left and left is equal-height or lower
        if (col != 0 && matrix[row][col] >= matrix[row][col - 1] && !visited.Contains($"{row}{col - 1}")) {
            visited.Add(key);
            var left = dfs (matrix, row, col - 1, ref dp, ref visited, ref finished);
            visited.Remove(key);
            res[0] = res[0] || left[0];
            res[1] = res[1] || left[1];
            if (res[0] && res[1]) {
                dp[row][col] = res;
                finished.Add(key);
                return res;
            }
        }
        // if has right and right is equal-height or lower
        if (col != matrix[0].Length - 1 && matrix[row][col] >= matrix[row][col + 1] && !visited.Contains($"{row}{col + 1}")) {
            visited.Add(key);
            var right = dfs (matrix, row, col + 1, ref dp, ref visited, ref finished);
            visited.Remove(key);
            res[0] = res[0] || right[0];
            res[1] = res[1] || right[1];
            if (res[0] && res[1]) {
                dp[row][col] = res;
                finished.Add(key);
                return res;
            }
        }

        if (!dp.ContainsKey (row)) {
            dp[row] = new Dictionary<int, bool[]> ();
        }
        if (!dp[row].ContainsKey (col)) {
            dp[row][col] = new bool[2];
        }
        
        dp[row][col] = res;
        finished.Add(key);
        return res;
    }
}
```
