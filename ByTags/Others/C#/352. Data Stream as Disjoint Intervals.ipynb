```
public class SummaryRanges {

    Queue<int[]> queue;
    
    /** Initialize your data structure here. */
    public SummaryRanges() {
        queue = new Queue<int[]>();
    }
    
    public void AddNum(int val) {
        if(queue.Count == 0){
            queue.Enqueue(new int[]{val, val});
        }else{
            var new_queue = new Queue<int[]>();
            var added = false;
            while(queue.Count != 0){
                var peek = queue.Dequeue();
                if(peek[1] == val - 1){
                    peek = new int[]{peek[0], val};
                    if(queue.Count != 0){
                        var next = queue.Peek();
                        if(next[0] == val + 1){
                            peek = new int[]{peek[0], next[1]};
                            queue.Dequeue();
                        }
                    }
                    new_queue.Enqueue(peek);
                    added = true;
                }else if(peek[0] == val + 1){
                    peek = new int[]{val, peek[1]};
                    new_queue.Enqueue(peek);
                    added = true;
                }else if(val < peek[0]){
                    if(!added){
                        new_queue.Enqueue(new int[]{val, val});
                    }
                    new_queue.Enqueue(peek);
                    added = true;
                }else{
                    if(peek[0] <= val && peek[1] >= val){
                        added = true;
                    }
                    new_queue.Enqueue(peek);
                }
                
            }
            if(!added){
                new_queue.Enqueue(new int[]{val, val});
            }
            queue = new_queue;
        }
        return;
    }
    
    public int[][] GetIntervals() {
        var temp = new Queue<int[]>(queue);
        var n = temp.Count;
        var res = new int[n][];
        var index = 0;
        while(index < n){
            res[index] = temp.Dequeue();
            index++;
        }
        return res;
    }
}

/**
 * Your SummaryRanges object will be instantiated and called as such:
 * SummaryRanges obj = new SummaryRanges();
 * obj.AddNum(val);
 * int[][] param_2 = obj.GetIntervals();
 */
```
