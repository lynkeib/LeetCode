```
public class Solution {
    
    HashSet<string> res;
    
    public IList<string> RemoveInvalidParentheses(string s) {
        res = new HashSet<string>();
        var invalid = numOfInvalidP(s);
        // Console.WriteLine($"{invalid.left}, {invalid.right}");
        RemoveP(new StringBuilder(s), 0, invalid.left, invalid.right);
        return res.ToList();
    }
    
    public void RemoveP(StringBuilder s, int start, int left, int right){
        // Console.WriteLine($"{left}, {right}");
        // Console.WriteLine(s.ToString());
        // Console.WriteLine();
        if(isValid(s.ToString()) && left == 0 && right == 0){
            res.Add(s.ToString());
            return;
        }
        for(int index = start; index < s.Length; index++){
            if(index != start && s[index] == s[index - 1]){
                continue;
            }
            if(right > 0 && s[index] == ')'){
                s.Remove(index, 1);
                RemoveP(s, index, left, right - 1);
                s.Insert(index, ')');
            }else if(left > 0 && s[index] == '('){
                s.Remove(index, 1);
                RemoveP(s, index, left - 1, right);
                s.Insert(index, '(');
            }
        }
        return;
    }
    
    public (int left, int right) numOfInvalidP(string s){
        int res = 0;
        int sum = 0;
        foreach(char c in s){
            if(c == '('){
                sum++;
            }else if(c == ')'){
                sum--;
            }
            if(sum < 0){
                res++;
                sum = 0;
            }
        }
        return (sum, res);
    }
    
    public bool isValid(string s){
        var count = 0;
        foreach(var c in s){
            if(c == '('){
                count ++;
            }else if(c == ')'){
                count--;
                if(count < 0){
                    return false;
                }
            }
        }
        return count == 0;
    }
}
```
