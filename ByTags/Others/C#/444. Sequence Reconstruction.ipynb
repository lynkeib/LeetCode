```
public class Solution {
    public bool SequenceReconstruction(int[] org, IList<IList<int>> seqs) {
        if(org.Length == 1){
            if(seqs.Count == 0){
                return false;
            }
            foreach(var l in seqs){
                if(l.Count != 1){
                    return false;
                }
                if(l[0] != org[0]){
                    return false;
                }
            }
            return true;
        }
        
        var graph = new Dictionary<int, HashSet<int>>();
        var inDegree = new Dictionary<int, int>();
        var q = new Queue<int>();
        var nodes = new HashSet<int>();
        var temp = new HashSet<int>(org);
        
        foreach(var l in seqs){
            if(l.Count == 1){
                if(!temp.Contains(l[0])){
                    return false;
                }
            }else{
                for(int i = 0; i < l.Count - 1; i++){
                    var parent = l[i];
                    var child = l[i + 1];
                    if(!temp.Contains(parent)){
                        return false;
                    }
                    if(!temp.Contains(child)){
                        return false;
                    }
                    if(!graph.ContainsKey(parent)){
                        graph[parent] = new HashSet<int>();
                        inDegree[parent] = 0;
                    }
                    if(!graph.ContainsKey(child)){
                        graph[child] = new HashSet<int>();
                        inDegree[child] = 0;
                    }
                    if(!graph[parent].Contains(child)){
                        graph[parent].Add(child);
                        inDegree[child]++;
                    }
                }
            }
            
        }
        foreach(var node in inDegree.Keys){
            if(inDegree[node] == 0){
                // Console.WriteLine($"{node}");
                q.Enqueue(node);
            }
        }
        var res = new List<int>();
        while(q.Count != 0){
            if(q.Count >= 2){
                return false;
            }
            var node = q.Dequeue();
            res.Add(node);
            foreach(var child in graph[node]){
                inDegree[child]--;
                if(inDegree[child] == 0){
                    q.Enqueue(child);
                }
            }
        }
        // foreach(var n in res) Console.Write($"{n} ");
        if(res.Count != org.Length){
            return false;
        }
        for(int i = 0; i < res.Count; i++){
            if(res[i] != org[i]){
                return false;
            }
        }
        return true;
    }
}
```
