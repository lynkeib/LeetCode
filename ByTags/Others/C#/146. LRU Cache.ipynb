# TLE

```
using System.Collections.Specialized;

public class LRUCache {

    private OrderedDictionary od;
    private int cap;

    public LRUCache (int capacity) {
        od = new OrderedDictionary ();
        cap = capacity;
    }

    public int Get (int key) {
        string k = key.ToString();
        if (od.Contains (k)) {
            int temp = (int) od[k];
            od.Remove (k);
            od.Insert (0, k, temp);
            return temp;
        }
        return -1;
    }

    public void Put (int key, int value) {
        string k = key.ToString();
        if (od.Contains (k)) {
            od.Remove (k);
            od.Insert (0, k, value);
        } else {
            od.Insert (0, k, value);
        }

        if (od.Count > cap) {
            od.RemoveAt(od.Count-1);
        }
    }
}

/**
 * Your LRUCache object will be instantiated and called as such:
 * LRUCache obj = new LRUCache(capacity);
 * int param_1 = obj.Get(key);
 * obj.Put(key,value);
 */
```

# LinkedListNode Wrapper

```
public class LRUCache {
    public class Cache {
        public int Key { get; set; }
        public int Value { get; set; }
    }

    private Dictionary<int, LinkedListNode<Cache>> dict;
    private LinkedList<Cache> list;
    int counter { get; set; }
    int cap { get; set; }

    public LRUCache (int capacity) {
        counter = 0;
        cap = capacity;
        dict = new Dictionary<int, LinkedListNode<Cache>> ();
        list = new LinkedList<Cache> ();
    }

    public int Get (int key) {
        if (dict.ContainsKey (key)) {
            list.Remove (dict[key]);
            list.AddFirst (dict[key]);
            return dict[key].Value.Value;
        }
        return -1;
    }

    public void Put (int key, int value) {
        if (dict.ContainsKey (key)) {
            list.Remove (dict[key]);
            list.AddFirst (dict[key]);
            dict[key].Value.Value = value;
            return;
        }
        if (counter >= cap) {
            dict.Remove (list.Last.Value.Key);
            list.RemoveLast ();
            counter --;
        }
        counter ++;
        Cache newCache = new Cache();
        newCache.Value = value;
        newCache.Key = key;
        LinkedListNode<Cache> newnode = new LinkedListNode<Cache> (newCache);
        dict[key] = newnode;
        list.AddFirst(newnode);
    }
}

/**
 * Your LRUCache object will be instantiated and called as such:
 * LRUCache obj = new LRUCache(capacity);
 * int param_1 = obj.Get(key);
 * obj.Put(key,value);
 */
```
