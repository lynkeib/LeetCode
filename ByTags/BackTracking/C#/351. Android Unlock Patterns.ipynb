```
public class Solution {
    
    int res;
    int m;
    int n;
    
    public int NumberOfPatterns(int m, int n) {
        this.m = m;
        this.n = n;
        
        for(int row = 0; row < 3; row++){
            for(int col = 0; col < 3; col++){
                var map = new int[3][];
                for(int r = 0; r < 3; r++){
                    map[r] = new int[3];
                }
                dfs(row, col, map, 0);
            }
        }
        
        return res;
    }
    
    public void dfs(int row, int col, int[][] path, int count){
        
        path[row][col] = 1;
        count++;
        
        if(count > n){
            path[row][col] = 0;
            return;
        }
        if(count >= m){
            // Console.WriteLine(count);
            res++;
            // Console.WriteLine($"{row} {col}");
        }
        // move up
        // move 1 unit
        if(row != 0 && path[row - 1][col] != 1){
            dfs(row - 1, col, path, count);
        }
        // move 2 units
        if(row == 2 && path[row - 1][col] == 1 && path[row - 2][col] != 1){
            dfs(row - 2, col, path, count);
        }
        
        // move down
        // move 1 unit
        if(row != 2 && path[row + 1][col] != 1){
            dfs(row + 1, col, path, count);
            // Console.WriteLine("??");
        }
        // move 2 units
        if(row == 0 && path[row + 1][col] == 1 && path[row + 2][col] != 1){
            dfs(row + 2, col, path, count);
        }
        
        // move right
        // move 1 unit
        if(col != 2 && path[row][col + 1] != 1){
            dfs(row, col + 1, path, count);
        }
        // move 2 units
        if(col == 0 && path[row][col + 1] == 1 && path[row][col + 2] != 1){
            dfs(row, col + 2, path, count);
        }
        
        // move left
        // move 1 unit
        if(col != 0 && path[row][col - 1] != 1){
            dfs(row, col - 1, path, count);
        }
        // move 2 units
        if(col == 2 && path[row][col - 1] == 1 && path[row][col - 2] != 1){
            dfs(row, col - 2, path, count);
        }
        
        // move bottom-right
        // move 1 unit
        if(row != 2 && col != 2 && path[row + 1][col + 1] != 1){
            dfs(row + 1, col + 1, path, count);
        }
        // move 2 units
        if(row == 0 && col == 0 && path[row + 1][col + 1] == 1 && path[row + 2][col + 2] != 1){
            dfs(row + 2, col + 2, path, count);
        }
        
        // move bottom-left
        // move 1 unit
        if(row != 2 && col != 0 && path[row + 1][col - 1] != 1){
            dfs(row + 1, col - 1, path, count);
        }
        // move 2 units
        if(row == 0 && col == 2 && path[row + 1][col - 1] == 1 && path[row + 2][col - 2] != 1){
            dfs(row + 2, col - 2, path, count);
        }
        
        // move top-right
        // move 1 unit
        if(row != 0 && col != 2 && path[row - 1][col + 1] != 1){
            dfs(row - 1, col + 1, path, count);
        }
        // move 2 units
        if(row == 2 && col == 0 && path[row - 1][col + 1] == 1 && path[row - 2][col + 2] != 1){
            dfs(row - 2, col + 2, path, count);
        }
        
        // move top-left
        // move 1 unit
        if(row != 0 && col != 0 && path[row - 1][col - 1] != 1){
            dfs(row - 1, col - 1, path, count);
        }
        // move 2 units
        if(row == 2 && col == 2 && path[row - 1][col - 1] == 1 && path[row - 2][col - 2] != 1){
            dfs(row - 2, col - 2, path, count);
        }
        
        // kight jump
        // bottom-right
        if(row + 2 < 3 && col + 1 < 3 && path[row + 2][col + 1] != 1){
            dfs(row + 2, col + 1, path, count);
        }
        if(row + 1 < 3 && col + 2 < 3 && path[row + 1][col + 2] != 1){
            dfs(row + 1, col + 2, path, count);
        }
        
        // bottom-left
        if(row + 2 < 3 && col - 1 > -1 && path[row + 2][col - 1] != 1){
            dfs(row + 2, col - 1, path, count);
        }
        if(row + 1 < 3 && col - 2 > -1 && path[row + 1][col - 2] != 1){
            dfs(row + 1, col - 2, path, count);
        }
        
        // top-right
        if(row - 2 > -1 && col + 1 < 3 && path[row - 2][col + 1] != 1){
            dfs(row - 2, col + 1, path, count);
        }
        if(row - 1 > -1 && col + 2 < 3 && path[row - 1][col + 2] != 1){
            dfs(row - 1, col + 2, path, count);
        }
        
        // top-left
        if(row - 2 > -1 && col - 1 > -1 && path[row - 2][col - 1] != 1){
            dfs(row - 2, col - 1, path, count);
        }
        if(row - 1 > -1 && col - 2 > -1 && path[row - 1][col - 2] != 1){
            dfs(row - 1, col - 2, path, count);
        }
        
        path[row][col] = 0;
        
        return;
    }
}
```
