```
public class Solution {
    public int CompareVersion(string version1, string version2) {
        var stack1 = new Stack<int>();
        var stack2 = new Stack<int>();
        var index1 = 0;
        var index2 = 0;
        while(index1 < version1.Length){
            var sum = 0;
            while(index1 < version1.Length && version1[index1] != '.'){
                if(sum == 0 && version1[index1] == '0'){
                    index1++;
                }else{
                    sum = sum * 10 + (version1[index1] - '0');
                    index1++;
                }
            }
            stack1.Push(sum);
            // Console.WriteLine(sum);
            index1++;
        }
        while(index2 < version2.Length){
            var sum = 0;
            while(index2 < version2.Length && version2[index2] != '.'){
                if(sum == 0 && version2[index2] == '0'){
                    index2++;
                }else{
                    sum = sum * 10 + (version2[index2] - '0');
                    index2++;
                }
            }
            stack2.Push(sum);
            // Console.WriteLine(sum);
            index2++;
        }
        while(stack1.Count != 0 && stack1.Peek() == 0){
            stack1.Pop();
        }
        while(stack2.Count != 0 && stack2.Peek() == 0){
            stack2.Pop();
        }
        var res = 0;
        if(stack1.Count > stack2.Count){
            res = 1;
        }
        while(stack1.Count > stack2.Count){
            stack1.Pop();
        }
        if(stack1.Count < stack2.Count){
            res = -1;
        }
        while(stack1.Count < stack2.Count){
            stack2.Pop();
        }
        while(stack1.Count != 0){
            var num1 = stack1.Pop();
            var num2 = stack2.Pop();
            // Console.WriteLine($"{num1}, {num2}");
            if(num1 > num2){
                res = 1;
            }else if(num1 < num2){
                res = -1;
            }
        }
        return res;
    }
}
```

```
public class Solution {
    public int CompareVersion(string version1, string version2) {
        int index1 = 0, index2 = 0;
        while (index1 < version1.Length || index2 < version2.Length)
        {
            int num1 = 0, num2 = 0;
            while (index1 < version1.Length && version1[index1] != '.')
            {
                num1 = num1 * 10 + (int)(version1[index1] - '0');
                index1++;
            }
            while (index2 < version2.Length && version2[index2] != '.')
            {
                num2 = num2 * 10 + (int)(version2[index2] - '0');
                index2++;
            }
            if (num1 > num2)
                return 1;
            else if (num1 < num2)
                return -1;
            else
            {
                index1++;
                index2++;
            }
        }
        return 0;
    }
}
```
