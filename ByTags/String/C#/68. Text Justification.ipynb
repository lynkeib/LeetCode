```
public class Solution {
    public IList<string> FullJustify (string[] words, int maxWidth) {
        var index = 0;
        var len = words.Length;
        var res = new List<string> ();
        while (index < len) {
            var this_line = new List<string> ();
            var word_count = 0;
            var space = 0;
            var word_space = 0;

            while (
                index < len &&
                (
                    (word_count != 0 && space + words[index].Length + 1 <= maxWidth) ||
                    (word_count == 0 && space + words[index].Length <= maxWidth)
                )
            ) {

                space = space + (word_count == 0 ? 0 : 1) + words[index].Length;
                word_count++;
                word_space += words[index].Length;
                this_line.Add (words[index]);
                // Console.WriteLine ($"Add {words[index]}, space left {maxWidth - space}");
                index++;

            }
            // Console.WriteLine ();
            var space_left = maxWidth - word_space;
            var intervals = word_count - 1;

            if (intervals == 0) {
                var this_string = this_line[0];
                while (space_left != 0) {
                    this_string += " ";
                    space_left--;
                }
                res.Add (this_string);
            } else {

                if (index == len) {
                    var this_string = this_line[0];
                    for (int i = 1; i < this_line.Count; i++) {
                        this_string = this_string + " " + this_line[i];
                        space_left--;
                    }
                    while (space_left != 0) {
                        this_string += " ";
                        space_left--;
                    }
                    res.Add (this_string);
                } else {
                    if (space_left % intervals == 0) {
                        var num_of_space = space_left / intervals;
                        var this_string = this_line[0];
                        for (int i = 1; i < this_line.Count; i++) {
                            for (int j = 0; j < num_of_space; j++) {
                                this_string += " ";
                            }
                            this_string += this_line[i];
                        }
                        res.Add (this_string);
                    } else {
                        var this_string = this_line[0];
                        for (int i = 1; i < this_line.Count; i++) {
                            if (space_left % intervals == 0) {
                                var num_of_space = space_left / intervals;
                                for (int j = 0; j < num_of_space; j++) {
                                    this_string += " ";
                                    space_left--;
                                }
                                this_string += this_line[i];
                            } else {
                                var num_of_space = space_left / intervals + 1;
                                for (int j = 0; j < num_of_space; j++) {
                                    if (space_left == 0) {
                                        break;
                                    }
                                    this_string += " ";
                                    space_left--;
                                }
                                this_string += this_line[i];
                            }
                            intervals--;
                        }
                        res.Add (this_string);
                    }
                }
            }
        }
        return res;
    }
}
```
