```
public class User {

    public IList<Tuple<int, int>> post { get; set; }
    public IList<int> follow { get; set; }

    public User (int userid) {
        post = new List<Tuple<int, int>> ();
        follow = new List<int> ();
        follow.Add (userid);
    }
}

public class Twitter {
    Dictionary<int, User> user;
    int time;
    /** Initialize your data structure here. */
    public Twitter () {
        user = new Dictionary<int, User> ();
        time = 0;
    }

    /** Compose a new tweet. */
    public void PostTweet (int userId, int tweetId) {
        if (!user.ContainsKey (userId)) {
            user[userId] = new User (userId);
        }
        user[userId].post.Insert (0, new Tuple<int, int> (time, tweetId));
        time += 1;
        return;
    }

    /** Retrieve the 10 most recent tweet ids in the user's news feed. Each item in the news feed must be posted by users who the user followed or by the user herself. Tweets must be ordered from most recent to least recent. */
    public IList<int> GetNewsFeed (int userId) {

        List<int> res = new List<int> ();

        if (!user.ContainsKey (userId)) return res;

        int[] indexes = new int[user[userId].follow.Count];
        int counter = 0;

        while (counter < 10) {
            int max_index = -1;
            int max_time = -1;
            bool hasleft = false;

            for (int i = 0; i < user[userId].follow.Count; i++) {
                int checkuser = user[userId].follow[i];
                int checkindex = indexes[i];
                if (checkindex < user[checkuser].post.Count) {
                    Tuple<int, int> post = user[checkuser].post[checkindex];
                    if (post.Item1 > max_time) {
                        max_index = i;
                        max_time = post.Item1;
                    }
                    hasleft = true;
                }
            }
            if (!hasleft) return res;
            int twitterId = user[user[userId].follow[max_index]].post[indexes[max_index]].Item2;
            res.Add (twitterId);
            indexes[max_index] += 1;
            counter += 1;
        }
        return res;
    }

    /** Follower follows a followee. If the operation is invalid, it should be a no-op. */
    public void Follow (int followerId, int followeeId) {
        if (!user.ContainsKey (followerId)) {
            user[followerId] = new User (followerId);
        }
        if (!user.ContainsKey (followeeId)) {
            user[followeeId] = new User (followeeId);
        }
        if (!user[followerId].follow.Contains (followeeId)) {
            user[followerId].follow.Add (followeeId);
        }

        return;
    }

    /** Follower unfollows a followee. If the operation is invalid, it should be a no-op. */
    public void Unfollow (int followerId, int followeeId) {
        if (user.ContainsKey (followerId)) {
            if (followerId != followeeId && user[followerId].follow.Contains (followeeId)) {
                user[followerId].follow.Remove (followeeId);
            }
        }
        return;
    }
}

/**
 * Your Twitter object will be instantiated and called as such:
 * Twitter obj = new Twitter();
 * obj.PostTweet(userId,tweetId);
 * IList<int> param_2 = obj.GetNewsFeed(userId);
 * obj.Follow(followerId,followeeId);
 * obj.Unfollow(followerId,followeeId);
 */
```
