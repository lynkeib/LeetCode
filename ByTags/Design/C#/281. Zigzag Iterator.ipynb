```
public class ZigzagIterator {
    IList<int> v1;
    IList<int> v2;
    int index;
    bool flap;

    public ZigzagIterator (IList<int> v1, IList<int> v2) {
        if (v1.Count > v2.Count) {
            this.v1 = v2;
            this.v2 = v1;
            flap = true;
        } else {
            this.v1 = v1;
            this.v2 = v2;
            flap = false;
        }
        index = 0;
    }

    public bool HasNext () {
        if (index / 2 < v1.Count) {
            return true;
        } else {
            return index - v1.Count < v2.Count;
        }
    }

    public int Next () {
        int res;
        if (flap) {
            if (index / 2 < v1.Count) {
                if (index % 2 == 0) {
                    res = v2[index / 2];
                } else {
                    res = v1[index / 2];
                }
            } else {
                res = v2[index - v1.Count];
            }
            index += 1;
        } else {
            if (index / 2 < v1.Count) {
                if (index % 2 == 0) {
                    res = v1[index / 2];
                } else {
                    res = v2[index / 2];
                }
            } else {
                res = v2[index - v1.Count];
            }
            index += 1;
        }
        return res;
    }
}

/**
 * Your ZigzagIterator will be called like this:
 * ZigzagIterator i = new ZigzagIterator(v1, v2);
 * while (i.HasNext()) v[f()] = i.Next();
 */
```
