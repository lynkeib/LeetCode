```
public class WordDictionary {
    public class Node {
        public char value { get; set; }
        public Dictionary<char, Node> children { get; set; }
        public bool end { get; set; }

        public Node (char value) {
            this.value = value;
            this.children = new Dictionary<char, Node> ();
            this.end = false;
        }
    }

    public Node head;

    /** Initialize your data structure here. */
    public WordDictionary () {
        head = new Node ('.');
    }

    /** Adds a word into the data structure. */
    public void AddWord (string word) {
        Node curr_head = head;
        int index = 0;
        while (index < word.Length && curr_head.children.ContainsKey(word[index])){
            curr_head = curr_head.children[word[index]];
            index ++;
        }
        while(index < word.Length){
            Node new_node = new Node(word[index]);
            curr_head.children[word[index]] = new_node;
            curr_head = new_node;
            index++;
        }
        curr_head.end = true;
        return;
    }

    /** Returns if the word is in the data structure. A word could contain the dot character '.' to represent any one letter. */
    public bool Search (string word) {
        return dfs(word, head.children);
    }

    public bool dfs(string word, Dictionary<char, Node> children){
        if(word.Length == 0) return false;

        if(word.Length == 1){
            if(word[0] == '.'){
                if(children.Count == 0) return false;
                foreach(char key in children.Keys){
                    if(children[key].end) return true;
                    return false;
                }
            }else{
                if(children.ContainsKey(word[0])){
                    return children[word[0]].end;
                }
                return false;
            }
        }
        if(word[0] == '.'){
            foreach(char key in children.Keys){
                if(dfs(word.Substring(1), children[key].children)){
                    return true;
                }
            }
            return false;
             
        }
        if(children.ContainsKey(word[0])) return dfs(word.Substring(1), children[word[0]].children);
        return false;
    }
}

/**
 * Your WordDictionary object will be instantiated and called as such:
 * WordDictionary obj = new WordDictionary();
 * obj.AddWord(word);
 * bool param_2 = obj.Search(word);
 */
```
