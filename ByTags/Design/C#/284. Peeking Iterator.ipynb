# Remembering the peeknum

```python
# Below is the interface for Iterator, which is already defined for you.
#
# class Iterator(object):
#     def __init__(self, nums):
#         """
#         Initializes an iterator object to the beginning of a list.
#         :type nums: List[int]
#         """
#
#     def hasNext(self):
#         """
#         Returns true if the iteration has more elements.
#         :rtype: bool
#         """
#
#     def next(self):
#         """
#         Returns the next element in the iteration.
#         :rtype: int
#         """

class PeekingIterator(object):
    def __init__(self, iterator):
        """
        Initialize your data structure here.
        :type iterator: Iterator
        """
        
        self.peeknum = iterator.next() if iterator.hasNext() else None;
        self.iterator = iterator;

    def peek(self):
        """
        Returns the next element in the iteration without advancing the iterator.
        :rtype: int
        """
        return self.peeknum;

    def next(self):
        """
        :rtype: int
        """
        res = self.peeknum;
        self.peeknum = self.iterator.next() if self.iterator.hasNext() else None;
        return res;

    def hasNext(self):
        """
        :rtype: bool
        """
        return self.peeknum is not None;

# Your PeekingIterator object will be instantiated and called as such:
# iter = PeekingIterator(Iterator(nums))
# while iter.hasNext():
#     val = iter.peek()   # Get the next element but not advance the iterator.
#     iter.next()         # Should return the same value as [val].
```

```python
// C# IEnumerator interface reference:
// https://docs.microsoft.com/en-us/dotnet/api/system.collections.ienumerator?view=netframework-4.8

class PeekingIterator {
    
    
    IEnumerator<int> iterator;
    int peek;
    bool hasnext;
    
    // iterators refers to the first element of the array.
    public PeekingIterator(IEnumerator<int> iterator) {
        // initialize any member here.
        this.iterator = iterator;
        peek = iterator.Current;
        hasnext = iterator.MoveNext();
    }
    
    // Returns the next element in the iteration without advancing the iterator.
    public int Peek() {
        return peek;
    }
    
    // Returns the next element in the iteration and advances the iterator.
    public int Next() {
        var res = peek;
        if(!hasnext){
            peek = int.MaxValue;
        }else{
            peek = iterator.Current;
            hasnext = iterator.MoveNext();
        }
        
        return res;
    }
    
    // Returns false if the iterator is refering to the end of the array of true otherwise.
    public bool HasNext() {
		return hasnext || peek != int.MaxValue;
    }
}
```
