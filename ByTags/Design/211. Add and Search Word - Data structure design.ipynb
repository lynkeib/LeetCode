```python
class Node:
    def __init__(self, value):
        self.val = value
        self.children = dict()
        self.end = False

    def AddChild(self, node):
        self.children[node.val] = node
        return
    
class WordDictionary(object):      
    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.head = Node('.')

    def addWord(self, word):
        """
        Adds a word into the data structure.
        :type word: str
        :rtype: None
        """
        head = self.head
        index = 0
        while index < len(word) and word[index] in head.children:
            head = head.children[word[index]]
            index += 1   
        while index != len(word):
            new_node = Node(word[index])
            head.AddChild(new_node)
            head = new_node
            index += 1
        head.end = True
        return

    def search(self, word):
        """
        Returns if the word is in the data structure. A word could contain the dot character '.' to represent any one letter.
        :type word: str
        :rtype: bool
        """
        return self.dfs(word, self.head.children)
    
    def dfs(self, word, children):
        if not children and word:
            return False
        if len(word) == 1:
            if word == '.':
                if not children:
                    return False
                for key, value in children.items():
                    if value.end:
                        return True
                return False
            else:
                if word not in children:
                    return False
                return children[word].end
        if word[0] == '.':
            for child in children:
                if self.dfs(word[1:], children[child].children):
                    return True
            return False
        else:
            if word[0] not in children:
                return False
            return self.dfs(word[1:], children[word[0]].children)
            


# Your WordDictionary object will be instantiated and called as such:
# obj = WordDictionary()
# obj.addWord(word)
# param_2 = obj.search(word)
```
