```
public class Solution {
    public int ShortestPathAllKeys(string[] grid) {
        
        var rows = grid.Length;
        var cols = grid[0].Length;
        
        var queue = new Queue<int>();
        var allKeys = 0;
        
        int[] dirs = {-1, 0, 1, 0, -1};
        
        // var state = new HashSet<int>();
        var state = new int[rows, cols, 64];
        
        for(int row = 0; row < rows; row++){
            for(int col = 0; col < cols; col++){
                var c = grid[row][col];
                if(c == '@'){
                    var s = ((row << 16) | (col << 8));
                    // state.Add(s);
                    state[row, col, 0] = 1;
                    queue.Enqueue(s);
                }else if(c >= 'a' && c <= 'f'){
                    allKeys |= (1 << (c - 'a'));
                }
            }
        }
        
        var count = 0;
        // var keys = 0;
        
        while(queue.Count != 0){
            var size = queue.Count;
            while(size != 0){
                var s = queue.Dequeue();
                var x = s >> 16;
                var y = (s >> 8) & 0xff;
                var keys = s & 0xff;
                if(keys == allKeys){
                    return count;
                }
                for(int i = 0; i < 4; i++){
                    var nx = x + dirs[i];
                    var ny = y + dirs[i + 1];
                    var nkeys = keys;
                    if(nx > -1 && nx < rows && ny > -1 && ny < cols){
                        var c = grid[nx][ny];
                        if(c == '#'){
                            continue;
                        }
                        if(c >= 'A' && c <= 'F'){
                            if(((1 << (c - 'A')) & keys) == 0){
                                continue;
                            }
                        }
                        if(c >= 'a' && c <= 'f'){
                            nkeys |= (1 << (c - 'a'));
                        }
                        if(state[nx, ny, nkeys] == 1){
                            continue;
                        }
                        queue.Enqueue((nx << 16) | (ny << 8) | nkeys);
                        state[nx, ny, nkeys] = 1;
                    }
                }
                size--;
            }
            count++;
        }
        return -1;
    }
}
```
