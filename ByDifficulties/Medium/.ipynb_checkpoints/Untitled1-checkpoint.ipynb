```python
class Solution(object):
    def wiggleMaxLength(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        if not nums:
            return 0
        res = 1
        state = 0 # 0 Begin, 1 Up, 2 Down
        p = []
        i = 0
        while i < len(nums) - 1:
            #print(i)
            if nums[i + 1] > nums[i] and state != 1:
                while i < len(nums) - 1 and nums[i + 1] > nums[i]:
                    i += 1
                #if i != len(nums) - 1 or nums[i] > nums[i - 1]:
                res += 1
                state = 1
                p.append((i, nums[i]))
            elif nums[i + 1] < nums[i] and state != 2:
                while i < len(nums) - 1 and nums[i + 1] < nums[i]:
                    i += 1
                #if i != len(nums) - 1 or nums[i] < nums[i - 1]:
                res += 1
                state = 2
                p.append((i, nums[i]))
            else:
                i += 1
                #state = 0
        #print(p)
        return res, p       
```

```python
class Solution(object):
    def canJump(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        if not nums:
            return False
        if len(nums) == 1:
            return True
        if nums[0] == 0:
            return False
        index = 0
        while index != len(nums) - 1:
            steps = nums[index]
            max_value = -1
            max_index = -1
            if index + steps >= len(nums) - 1:
                return True
            for i in range(index + 1, min(len(nums), index + steps + 1)):
                if nums[i] + i >= max_value:
                    max_value = nums[i] + i
                    max_index = i
            if max_value - max_index == 0:
                if max_index == len(nums) - 1:
                    return True
                return False
            if max_index == len(nums) - 1:
                return True
            index = max_index
```

```python
class Solution(object):
    def calculate(self, s):
        """
        :type s: str
        :rtype: int
        """
        stack = []
        pre_op, num = "+", 0
        s += '+0' #ensure that if the last op is "*" or "/", we will process them in the end of loop
        for i in range(len(s)):
            if s[i] >= '0' and s[i] <= '9':
                num = num * 10 + int(s[i])
            else:
                if pre_op == '+':
                    stack.append(num)
                elif pre_op == '-':
                    stack.append(-num)
                elif pre_op == '*':
                    stack.append(stack.pop() * num)
                else:
                    stack.append(stack.pop() / num)
                num = 0
                pre_op = s[i]
        return sum(stack)
```

```python
class Solution(object):
    def findMinArrowShots(self, points):
        """
        :type points: List[List[int]]
        :rtype: int
        """
        if not points:
            return 0
        if len(points) == 1:
            return 1
        points = sorted(points, key = lambda a: a[0])
        res = 0
        index = 0
        while index < len(points):
            left, right = points[index]
            #print(left, right)
            res += 1
            index += 1
            while index < len(points) and points[index][0] <= right:
                #print(points)
                #print(points[0][0])
                right = min(right, points[index][1])
                index += 1
        return res
```
