```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def lowestCommonAncestor(self, root, p, q):
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """
        parent = dict()
        level = dict()
        self.find_parent(root, parent)
        self.find_level(root, level, 0)
        if level[p] == level[q]:
            while p != q:
                p = parent[p]
                q = parent[q]
            return q
        elif level[p] > level[q]:
            while level[p] != level[q]:
                p = parent[p]
            while p != q:
                p = parent[p]
                q = parent[q]
            return q
        else:
            while level[p] != level[q]:
                q = parent[q]
            while p != q:
                p = parent[p]
                q = parent[q] 
            return q
    
    def find_parent(self, root, parent):
        if not root:
            return
        if root.left:
            parent[root.left] = root
            self.find_parent(root.left, parent)
        if root.right:
            parent[root.right] = root
            self.find_parent(root.right, parent)
        return 
        
    def find_level(self, root, level, curr_level):
        if not root:
            return
        level[root] = curr_level
        self.find_level(root.left, level, curr_level + 1)
        self.find_level(root.right, level, curr_level + 1)
        return 
```

```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution(object):
    def lowestCommonAncestor(self, root, p, q):
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """
        #print(p.val)
        path_p = self.find_path(root, p, [])
        path_q = self.find_path(root, q, [])
        #print(path_p, path_q)
        
        #print(path_p)
        for i in range(1, min(len(path_p), len(path_q))):
            if path_p[i] != path_q[i]:
                return path_p[i - 1]
            if i == min(len(path_p), len(path_q)) - 1:
                return path_p[i]
        return path_p[0]
        
    def find_path(self, root, target, path):
        if not root:
            return
        if root == target:
            return path + [root]
        left = self.find_path(root.left, target, path + [root])
        if left:
            return left
        right = self.find_path(root.right, target, path + [root])
        if right:
            return right
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def lowestCommonAncestor(self, root, p, q):
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """
        return self.dfs(root, p, q)
        
    def dfs(self, root, target1, target2):
        if not root:
            return
        if root == target1 or root == target2:
            return root
        left = self.dfs(root.left, target1, target2)
        right = self.dfs(root.right, target1, target2)
        if left and right:
            return root
        elif left:
            return left
        else:
            return right
```
