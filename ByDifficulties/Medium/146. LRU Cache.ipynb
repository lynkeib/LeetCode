```python
class Node(object):
    
    def __init__(self, key, val):
        self.val = val
        self.key = key
        self.next = None
        self.prev = None

class LRUCache(object):

    def __init__(self, capacity):
        """
        :type capacity: int
        """
        self.capacity = capacity
        self.dict = dict()
        self.dummyhead = Node(None, None)
        self.dummytail = Node(None, None)
        self.dummyhead.next = self.dummytail
        self.dummytail.prev = self.dummyhead

    def get(self, key):
        """
        :type key: int
        :rtype: int
        """
        #print('get', key)
        if key in self.dict:
            self.put_front(self.dict[key])
            return self.dict[key].val
        #self.print_list()
        return -1

    def put(self, key, value):
        """
        :type key: int
        :type value: int
        :rtype: None
        """
        #print('put', key, value)
        if key in self.dict:
            self.dict[key].val = value
            self.put_front(self.dict[key])
            #self.print_list()
            return 
        if len(self.dict) == self.capacity:
            self.del_from_back()
        # self.print_list()
        self.dict[key] = Node(key, value)
        self.put_front(self.dict[key])
        #self.print_list()
        return 
    
    def put_front(self, node):
        if node.prev and node.next:
            node.prev.next = node.next
            node.next.prev = node.prev
        temp = self.dummyhead.next
        node.next = temp
        temp.prev = node
        self.dummyhead.next = node
        node.prev = self.dummyhead
        return 
    
    def del_from_back(self):
        node = self.dummytail.prev
        if not node.val:
            return
        node.prev.next = self.dummytail
        self.dummytail.prev = node.prev
        del self.dict[node.key]
        return 
    
    def print_list(self):
        node = self.dummyhead
        res = []
        while node:
            res.append(node.key)
            node = node.next
        print(res)
        return 

# Your LRUCache object will be instantiated and called as such:
# obj = LRUCache(capacity)
# param_1 = obj.get(key)
# obj.put(key,value)
```
