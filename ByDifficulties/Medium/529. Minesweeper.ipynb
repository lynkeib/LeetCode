```python
class Solution(object):
    def updateBoard(self, board, click):
        """
        :type board: List[List[str]]
        :type click: List[int]
        :rtype: List[List[str]]
        """
        self.click(click[0], click[1], board)
        return board
    
    def click(self, x, y, board):
        if board[x][y] == 'M':
            board[x][y] = 'X'
            return
        self.dfs(x, y, board)
    
    def howmanymine(self, x, y, board):
        count = 0
        #down
        if x+1<len(board):
            count += (board[x+1][y] in "MX")
            #down-left
            if y-1>-1:
                count += (board[x+1][y-1] in "MX")
            #down-right
            if y+1<len(board[0]):
                count += (board[x+1][y+1] in "MX")
        #up
        if x-1>-1:
            count += (board[x-1][y] in "MX")
            #up-left
            if y-1>-1:
                count += (board[x-1][y-1] in "MX")
            #up-right
            if y+1<len(board[0]):
                count += (board[x-1][y+1] in "MX")
        #left
        if y-1>-1:
                count += (board[x][y-1] in "MX")
        #right
        if y+1<len(board[0]):
            count += (board[x][y+1] in "MX")
        return count

    def dfs(self, x, y, board):
        if board[x][y] in "B1MX":
            return 
        mine_counter = self.howmanymine(x, y, board)
        if mine_counter:
            board[x][y] = str(mine_counter)
            return 
        board[x][y] = 'B'
        #down
        if x+1<len(board):
            self.dfs(x+1,y,board)
            #down-left
            if y-1>-1:
                self.dfs(x+1,y-1,board)
            #down-right
            if y+1<len(board[0]):
                self.dfs(x+1,y+1,board)
        #up
        if x-1>-1:
            self.dfs(x-1,y,board)
            #up-left
            if y-1>-1:
                self.dfs(x-1,y-1,board)
            #up-right
            if y+1<len(board[0]):
                self.dfs(x-1,y+1,board)
        #left
        if y-1>-1:
            self.dfs(x,y-1,board)
        #right
        if y+1<len(board[0]):
            self.dfs(x,y+1,board)
        return 
        
        
```
