```python
class Solution(object):
    def minWindow(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: str
        """
        if not s or not t:
            return ""
        if len(t) > len(s):
            return ""
        # if len(t) == len(s):
        #     return ""
        ## construc dict_t
        dict_t = collections.defaultdict(int)
        for i in t:
            dict_t[i] += 1
        ## initialize dict_s
        dict_s = collections.defaultdict(int)
        
        slow = 0
        
        res = ""
        
        for fast in range(len(s)):
            dict_s[s[fast]] += 1
            ## if slow < fast, we want to move slow as far as possible
            while slow < fast:
                # if chr at slow not in t
                if s[slow] not in dict_t:
                    slow += 1
                # if the number of chr at slow is large enough
                elif dict_s[s[slow]] > dict_t[s[slow]]:
                    dict_s[s[slow]] -= 1
                    slow += 1
                else:
                    break
            # check if the current window is ok
            if self.t_in(dict_s, dict_t):
                length = fast - slow + 1
                if res == "" or length < len(res):
                    res = s[slow:fast + 1]
        return res
        
    def t_in(self, s, t):
        for key, value in t.items():
            if key not in s:
                return False
            elif s[key] < value:
                    return False
        return True
```
