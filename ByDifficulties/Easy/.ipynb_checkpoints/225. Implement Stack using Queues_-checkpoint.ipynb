```python
class queue(object):
    def __init__(self):
        self.nums = []
        self.size = 0
    
    def push_to_back(self, num):
        self.nums.append(num)
        self.size += 1
    
    def pop_from_front(self):
        if self.nums:
            self.size -= 1
            return self.nums.pop(0)
        return None
    
    def size(self):
        return self.size
    
    def is_empty(self):
        return self.size == 0
    
    def front(self):
        if self.nums:
            return self.nums[0]
        return None

class MyStack(object):

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.queue = queue()

    def push(self, x):
        """
        Push element x onto stack.
        :type x: int
        :rtype: None
        """
        tempqueue = queue()
        tempqueue.push_to_back(x)
        while self.queue.size != 0:
            tempqueue.push_to_back(self.queue.pop_from_front())
        self.queue = tempqueue

    def pop(self):
        """
        Removes the element on top of the stack and returns that element.
        :rtype: int
        """
        if not self.queue.is_empty():
            return self.queue.pop_from_front()
        return None
        

    def top(self):
        """
        Get the top element.
        :rtype: int
        """
        return self.queue.front()
        

    def empty(self):
        """
        Returns whether the stack is empty.
        :rtype: bool
        """
        return self.queue.is_empty()


# Your MyStack object will be instantiated and called as such:
# obj = MyStack()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.top()
# param_4 = obj.empty()
```
