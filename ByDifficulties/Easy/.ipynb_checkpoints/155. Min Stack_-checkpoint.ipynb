```python
class MinStack(object):

    def __init__(self):
        """
        initialize your data structure here.
        """
        self._data = []
        self._size = 0
        self._min = []
        

    def push(self, x):
        """
        :type x: int
        :rtype: None
        """
        if self._size:
            if x < self._min[-1]:
                self._min.append(x)
            else:
                self._min.append(self._min[-1])
            self._data.append(x)
            self._size += 1
        else:
            self._data.append(x)
            self._min.append(x)
            self._size += 1
        
    def pop(self):
        """
        :rtype: None
        """
        if self._size:
            self._size -= 1
            self._min.pop()
            return self._data.pop()
        return None
        

    def top(self):
        """
        :rtype: int
        """
        if self._size:
            return self._data[-1]
        return None
        

    def getMin(self):
        """
        :rtype: int
        """
        if self._size:
            return self._min[-1]
        return None


# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(x)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
```
